## Step 4: Initial sanitization

We will eliminate:
* __location__, since the same data are already given as __latitude__ and __longitude__; and,
* __collision id__, since we do not require a primary key to use as a foreign key in any other table.

We will rename:
* __crash date__ to "date," since this is already a table of collisions; and
* __crash time__ to "time," for the same reason.

From observing the data, we notice that "EXPRESSWAY" is sometimes abbreviated "EXPY" and "PARKWAY" is sometimes abbreviated "PKWY." Before evaluating streets, we need to sanitize the __on street name__, __cross street name__ and __off street name__ columns.

Finally, we covert the __crash date__ field to `datetime` before using the field in our analysis.

```python
from pandas import to_datetime

table = table.drop([ 'LOCATION', 'COLLISION_ID' ], axis = 1)
table = table.rename(columns = { 'CRASH DATE': 'DATE', 'CRASH_TIME': 'TIME' })

# https://stackoverflow.com/questions/28986489/how-to-replace-text-in-a-string-column-of-a-pandas-dataframe

for column in [ 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME' ]:
    table[column] = table[column].str.replace(" PARKWAY", " PKWY")
    table[column] = table[column].str.replace(" EXPRESSWAY", " EXPY")

table['DATE'] = to_datetime(table['DATE'], format = '%m/%d/%y')
```

## Step 7: Summary statistics for number of persons injured

We can compute summary statistics for the number of persons injured in New York City.

```python
def summary_statistics(column):
    print("Mean:", column.mean())
    print("Min:", column.min())
    print("25th percentile:", column.quantile(0.25))
    print("Median:", column.median())
    print("7th percentile:", column.quantile(0.75))
    print("Max:", column.max())

summary_statistics(table['NUMBER OF PERSONS INJURED'])
```

We can also compute summary statistics for two selected boroughs.

```python
for borough in [ 'MANHATTAN', 'QUEENS' ]:
    column = table[table['BOROUGH'] == borough]['NUMBER OF PERSONS INJURED']

    print(borough)
    summary_statistics(column)
    print()
```

We notice that at the 75th percentile, the number of persons injured in Manhattan is 0. This implies that Manhattan is quite safe in terms of injuries from traffic accidents. Relative to Queens, Manhattan's mean number of persons injured is less (5, compared with 7). In Queens, the 75th-percentile number of persons injured is 1.0.

These results are within expectation since vehicles are often stuck in slow traffic in Manhattan, which may reduce the rate of injuries. Also, driving is uncommon in Manhattan, which may influence the results.

## Step 10: Which month did the most accidents occur?

Having converted the __crash date__ column to a `datetime` field, we can easily extract and visualize the months.

```python
%pip install matplotlib

from matplotlib.pyplot import bar, show

months = table['DATE'].dt.month.value_counts()

bar(months.index, months.values)
show()
```

The month with the most accidents in 2020 is January. Although this is not what we might expect from a typical year in New York City, these data are confounded by the implementation of the COVID-19 shutdown in March 2020.
